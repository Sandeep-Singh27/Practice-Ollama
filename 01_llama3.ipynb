```python
import requests
import json

payload = {
    "model":"llama3:8b",
    "prompt":"how to write print statement in python?"
}

response = requests.post("http://localhost:11434/api/generate",json=payload, stream=True)

count = 0
for i in response.iter_lines():
    print(json.loads((i.decode())).get("response"), end=" ")
    if count%10 == 0:
        print()
    count += 1
```
